```python
#Memanggil library/package yang digunakan
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctr
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Fuzzifikasi

## Variabel input ( Antecedent):
1. Suhu : Dingin, Sejuk, Panas.
2. Umur : Anak, Peranggang, Dewasa
3. Populasi : Sedikit, Sedang, Banyak.

## Variabel Output (Consequent):
Pemanas : Nyalakan Pemanas, Nyalakan Pendingin.

```python
# Deklarasi Input dan Output
x_suhu = ctr.Antecedent(np.arange(27,34,1), 'x_suhu')
x_umur = ctr.Antecedent(np.arange(1,34,1), 'x_umur')
x_populasi = ctr.Antecedent(np.arange(0,3501,1), 'x_populasi')
x_pemanas = ctr.Consequent(np.arange(27,34,1), 'x_pemanas')
```

# Fungsi Keanggotaan
<h2>1. Suhu sebagai Variabel Input  </h2>

Fungsi Keanggotaan Suhu

$Suhu [Dingin] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \geq 30 \\
       {30-x \over 30-28} & \mbox{; } 28 ≤ x ≤ 30 \\
       1 & \mbox{; } x ≤ 28
    \end{array}
\right.$

$Suhu [Sejuk] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 29 \ atau \ x\geq 31 \\
       {x-29 \over 30-29} & \mbox{; } 30 ≤ x ≤ 29 \\
       1 & \mbox{; } x = 30\\
       {30-x \over 31-30} & \mbox{; } 30 ≤ x ≤ 31 \\
    \end{array}
\right.$

$Suhu [Panas] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 30  \\
       {x-30 \over 32-30} & \mbox{; } 30 ≤ x ≤ 32 \\
       1 & \mbox{; } x\geq 32\\
    \end{array}
\right.$


<h2>2. Umur sebagai Variabel Input  </h2>

Fungsi Keanggotaan Umur

$Umur [Anak] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \geq 4 \\
       {4-x \over 4-2} & \mbox{; } 2 ≤ x ≤ 4 \\
       1 & \mbox{; } x ≤ 2
    \end{array}
\right.$

$Umur [Peranggang] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 3 \ atau \ x\geq 9 \\
       {x-3 \over 5-3} & \mbox{; } 3 ≤ x ≤ 5 \\
       1 & \mbox{; } 5 ≤ x ≤ 7 \\
       {7-x \over 9-7} & \mbox{; } 7 ≤ x ≤ 9 \\
    \end{array}
\right.$

$Umur [Dewasa] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 8  \\
       {x-8 \over 10-8} & \mbox{; } 8 ≤ x ≤ 10 \\
       1 & \mbox{; } x\geq 33\\
    \end{array}
\right.$


<h2>3. Populasi sebagai Variabel Input  </h2>

Fungsi Keanggotaan Populasi

$Populasi [Sedikit] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \geq 2000 \\
       {2000-x \over 2000-1000} & \mbox{; } 1000 ≤ x ≤ 2000 \\
       1 & \mbox{; } x ≤ 1000
    \end{array}
\right.$

$Populasi [Sedang] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 1500 \ atau \ x\geq 2500 \\
       {x-1500 \over 2000-1500} & \mbox{; } 1500 ≤ x ≤ 2000 \\
       1 & \mbox{; } x = 2000 \\
       {2000-x \over 2500-2000} & \mbox{; } 2000 ≤ x ≤ 2500 \\
    \end{array}
\right.$

$Populasi [Banyak] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 2000  \\
       {x-2000 \over 3000-2000} & \mbox{; } 2000 ≤ x ≤ 3000 \\
       1 & \mbox{; } x\geq 3500\\
    \end{array}
\right.$

<h2>Pemanas sebagai Variabel Output  </h2>

Fungsi Keanggotaan Pemanas

$Pemanas [Nyalakan Pemanas] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \geq 31 \\
       {31-x \over 31-29} & \mbox{; } 29 ≤ x ≤ 31 \\
       1 & \mbox{; } x ≤ 29
    \end{array}
\right.$

$Pemanas [Nyalakan Pendingin] =
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x ≤ 30  \\
       {x-30 \over 32-30} & \mbox{; } 30 ≤ x ≤ 32 \\
       1 & \mbox{; } x\geq 32\\
    \end{array}
\right.$

```python
#Deklarasikan masing-masing parameter dari variabel
x_suhu ['dingin'] = fuzz.trapmf(x_suhu.universe,(0,0,28,30))
x_suhu ['sejuk'] = fuzz.trimf(x_suhu.universe,(29,30,31))
x_suhu ['panas'] = fuzz.trapmf(x_suhu.universe,(30,32,34,34))
x_suhu.view();
```

```python
x_umur ['Anak'] = fuzz.trapmf(x_umur.universe,(0,0,2,4))
x_umur ['Peranggang'] = fuzz.trapmf(x_umur.universe,(3, 5, 7, 9))
x_umur ['Dewasa'] = fuzz.trapmf(x_umur.universe,(8, 10, 33, 33))
x_umur.view();
```

```python
x_populasi ['Sedikit'] = fuzz.trapmf(x_populasi.universe,(0,0,1000,2000))
x_populasi ['Sedang'] = fuzz.trimf(x_populasi.universe,(1500,2000,2500))
x_populasi ['Banyak'] = fuzz.trapmf(x_populasi.universe,(2000,3000,3500,3500))
x_populasi.view();
```

```python
x_pemanas ['Nyalakan Pemanas'] = fuzz.trapmf(x_pemanas.universe,(0, 0, 29, 31))
x_pemanas ['Nyalakan Pendingin'] = fuzz.trapmf(x_pemanas.universe,(30, 32, 33, 33))
x_pemanas.view();
```

# 2. Fuzzy Inference

<h2> Rule Base </h2>

Dalam kasus fuzzy ini terdapat  27 rules yang  didapatkan.

```python
rule1 = ctr.Rule(x_suhu['dingin'] & x_umur['Anak']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pemanas'])
rule2 = ctr.Rule(x_suhu['dingin'] & x_umur['Anak']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pemanas'])
rule3 = ctr.Rule(x_suhu['dingin'] & x_umur['Anak']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pemanas'])
rule4 = ctr.Rule(x_suhu['dingin'] & x_umur['Peranggang']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pemanas'])
rule5 = ctr.Rule(x_suhu['dingin'] & x_umur['Peranggang']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pemanas'])
rule6 = ctr.Rule(x_suhu['dingin'] & x_umur['Peranggang']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pemanas'])
rule7 = ctr.Rule(x_suhu['dingin'] & x_umur['Dewasa']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pemanas'])
rule8 = ctr.Rule(x_suhu['dingin'] & x_umur['Dewasa']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pemanas'])
rule9 = ctr.Rule(x_suhu['dingin'] & x_umur['Dewasa']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pemanas'])
rule10 = ctr.Rule(x_suhu['sejuk'] & x_umur['Anak']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pemanas'])
rule11 = ctr.Rule(x_suhu['sejuk'] & x_umur['Anak']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pemanas'])
rule12 = ctr.Rule(x_suhu['sejuk'] & x_umur['Anak']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pemanas'])
rule13 = ctr.Rule(x_suhu['sejuk'] & x_umur['Peranggang']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pemanas'])
rule14 = ctr.Rule(x_suhu['sejuk'] & x_umur['Peranggang']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pemanas'])
rule15 = ctr.Rule(x_suhu['sejuk'] & x_umur['Peranggang']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pendingin'])
rule16 = ctr.Rule(x_suhu['sejuk'] & x_umur['Dewasa']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pendingin'])
rule17 = ctr.Rule(x_suhu['sejuk'] & x_umur['Dewasa']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pendingin'])
rule18 = ctr.Rule(x_suhu['sejuk'] & x_umur['Dewasa']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pendingin'])
rule19 = ctr.Rule(x_suhu['panas'] & x_umur['Anak']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pendingin'])
rule20 = ctr.Rule(x_suhu['panas'] & x_umur['Anak']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pendingin'])
rule21 = ctr.Rule(x_suhu['panas'] & x_umur['Anak']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pendingin'])
rule22 = ctr.Rule(x_suhu['panas'] & x_umur['Peranggang']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pendingin'])
rule23 = ctr.Rule(x_suhu['panas'] & x_umur['Peranggang']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pendingin'])
rule24 = ctr.Rule(x_suhu['panas'] & x_umur['Peranggang']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pendingin'])
rule25 = ctr.Rule(x_suhu['panas'] & x_umur['Dewasa']  & x_populasi['Sedikit'], x_pemanas['Nyalakan Pendingin'])
rule26 = ctr.Rule(x_suhu['panas'] & x_umur['Dewasa']  & x_populasi['Sedang'], x_pemanas['Nyalakan Pendingin'])
rule27 = ctr.Rule(x_suhu['panas'] & x_umur['Dewasa']  & x_populasi['Banyak'], x_pemanas['Nyalakan Pendingin'])
```

```python
rule1.view();
```

<h2> Metode Mamdani </h2>

```python
pemanas_rule = ctr.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
pemanas = ctr.ControlSystemSimulation(pemanas_rule)
```

<h2> 3. Defuzzifikasi </h2>

```python
# Membuat interupsi yang meminta inputan dari keyboard
suhu = input('Suhu adalah : ')
umur = input('Umur adalah : ')
populasi = input('Populasi adalah : ')
```

Pada kandang ayam broiler dengan suhu 28 derajat , umur ayam adalah 2 hari dan populasi pada kandang sebanyak 2700 ekor. Tentukan Output nya!

1. $Input.SUHU[28]=
\left\{
    \begin{array}{ll}
       {30-28 \over 30-28}  & = 1 \\
    \end{array}
\right.$



2. $Input.UMUR[2] =
\left\{
    \begin{array}{ll}
       {4-2 \over 4-2} & = 1 \\
    \end{array}
\right.$
    

3. $Input.POPULASI[2700] =
\left\{
    \begin{array}{ll}
       {2700-2000 \over 3000-2000} & = 0.7 \\
    \end{array}
\right.$


```python
pemanas.input['x_suhu'] = float(suhu) # Mengubah tipe data integer ke float
pemanas.input['x_umur'] = float(umur)
pemanas.input['x_populasi'] = float(populasi)
```

```python
pemanas.compute()
```

```python
print(pemanas.output['x_pemanas'])
```

```python
x_suhu.view(sim=pemanas)
x_umur.view(sim=pemanas)
x_populasi.view(sim=pemanas)
plt.show
```
